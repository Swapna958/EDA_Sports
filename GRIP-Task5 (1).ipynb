```python
plt.figure(figsize=(15,8))
sns.countplot(x='season', data=matches)
plt.show()
```

## Here In The Year 2011,2012 And 2013 There Was Maximum Number of Matches Has Been Played.

```python
matches.groupby('season')['team1'].nunique().plot(kind = 'bar', figsize=(15,10))
plt.title("Number of teams participated each season ",fontsize=18,fontweight="bold")
plt.ylabel("Count of teams", size = 25)
plt.xlabel("Season", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

## Here We Can See 10 Teams Played In 2011 And 9 Teams Each In 2012 and 2013. 

```python
Final_data['win_by']=np.where(Final_data['win_by_runs']>0,'Bat first','Bowl first')
```

```python
plt.figure(figsize = (18,10))
sns.countplot('season',hue='win_by',data=Final_data,palette='Set1')
plt.title("Numbers of matches won by batting and bowling first ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

## Here We Can See That Maximum Number of Matches Win By Chossing Bowling First.

```python
Toss=Final_data.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['#FFBF00', '#DD7596']
plt.figure(figsize = (15,10))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Toss Result',fontsize=20)
plt.axis('equal')
plt.show()
```

## Here We Can See That After Winning The Toss, The Maximum Probability Is Chossing A Field First And It Is 61%.

```python
Final_data['toss_win_game_win'] = np.where((Final_data.toss_winner == Final_data.winner),'Yes','No')
plt.figure(figsize = (15,10))
sns.countplot('toss_win_game_win', data=Final_data, hue = 'toss_decision')
plt.title("How Toss Decision Affects Match Result", fontsize = 20)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.xlabel("Winning Toss And Winning Match", fontsize = 20)
plt.ylabel("Frequency", fontsize = 20)
```

```python
plt.figure(figsize = (15,10))
sns.countplot('season',hue='toss_decision',data=Final_data,palette='Set2')
plt.title("Numbers of Matches Won By Toss Result ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

## Here We Can See That In Recent Years There Is High Rate of Winning When The Team Choosing Field First After Winning The Toss.

## Here Mumbai Indians Is The Leading Team which Wins The Most Number of IPL Trophies. Also Chennai Super Kings Is At Second Position.

```python
temp_df = pd.melt(Final_data, id_vars=['id','season'], value_vars=['team1', 'team2'])

plt.figure(figsize=(15,10))
sns.countplot(x='value', data=temp_df)
plt.xlabel("Teams",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.xticks(rotation='vertical')
plt.show()
```

```python
Number_of_fours = Final_data[Final_data['batsman_runs'] == 4]
Number_of_fours.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python
Number_of_six = Final_data[Final_data['batsman_runs'] == 6]
Number_of_six.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

```python
plt.figure(figsize=(18,10))
ax=sns.countplot(Final_data.dismissal_kind)
plt.title("Dismissals in IPL",fontsize=20)
plt.xlabel("Dismissals Kind",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (25,10))
sns.countplot('toss_winner', data = Final_data, hue = 'toss_decision')
plt.title("Teams decision to bat first or second after winning toss", size = 30)
plt.xticks(size = 15, rotation=50)
plt.yticks(size = 15)
plt.xlabel("Toss Winner", size = 25)
plt.ylabel("Count", size = 25)
```

```python
bowler_wickets = Final_data.groupby('bowler').aggregate({'ball': 'count', 'total_runs': 'sum', 'player_dismissed' : 'count'})
bowler_wickets.columns = ['runs','balls','wickets']
TopBowlerBywickets = bowler_wickets.sort_values(['wickets'], ascending=False)[:20]
TopBowlerBywickets
```

```python
Man_of_The_Match = Final_data['player_of_match'].value_counts()
Man_of_The_Match.head(10).plot(kind = 'bar',figsize=(12,8), fontsize=15, color='green')
plt.title("Top 10 Players With Most Man of The Match Awards",fontsize=20)
plt.ylabel("Frequency", size = 20)
plt.xlabel("Players", size = 20) 
```

## Here CH Gayle And AB de Villiers Are Two Players Who Wins The Most Number of Man of The Match Awards In IPL.

```python
Final_data.groupby('bowler')['player_dismissed'].count().sort_values(ascending = False).head(10).plot(kind = 'bar', 
                                                color = 'purple', figsize = (15,10))
plt.title("Top Wicket Takers of IPL", fontsize = 25,)
plt.xlabel("Bowler Name", size = 20)
plt.ylabel("Total Wickets Taken", size = 20)
plt.xticks(size = 12)
plt.yticks(size = 12)
```

## Here We Can See That Lasith Malinga Is The Highest Wicket Taker Bowler In The IPL.

```python
Final_data.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'orange',
                                                                                            figsize = (15,5))
plt.title("Top Run Getters of IPL", fontsize = 25)
plt.xlabel("Batsmen Name", size = 20)
plt.ylabel("Total Runs Scored", size = 20)
plt.xticks(size = 12)
plt.yticks(size = 12)
```
